```python
# prompt: crear un arreglo con la variable datos, con datos sinteticos, usando numpy donde la columna 0, sea la edad y la columna 1 sea la nota de IA que van de 0 a 5,

import numpy as np

# Crear un arreglo con datos sintÃ©ticos usando NumPy
num_samples = 20
edades = np.random.randint(18, 25, size=num_samples)
notas_ia = np.random.uniform(0, 5, size=num_samples).round(1)  # Notas de IA de 0 a 5

datos = np.column_stack((edades, notas_ia))

datos
```

```python
# prompt: calcular el promedio de la columna 1 de de datos

import numpy as np
# Calcular el promedio de la columna 1 (edades) del arreglo 'datos'
promedio_edades = np.mean(datos[:, 1])

print(f"El promedio de las edades es: {promedio_edades}")
```

```python
# prompt: crear una lista de numpy con 20 nombres de alumnos sinteticos, es una variable llamada nombres, inventar los nombres de la lista

import numpy as np

nombres = np.array([
    "Alice", "Bob", "Charlie", "David", "Emily",
    "Frank", "Grace", "Henry", "Isabella", "Jack",
    "Katherine", "Liam", "Mia", "Noah", "Olivia",
    "Peter", "Quinn", "Ryan", "Sophia", "Thomas"
])
nombres
```

```python
# prompt: crear un data frame llamado alumnos donde integramos el arreglo nombres, edades, notas de IA

import pandas as pd
alumnos = pd.DataFrame({
    'Nombre': nombres,
    'Edad': edades,
    'Nota_IA': notas_ia
})
alumnos
```
